```python
import torch
from torch.utils.data import Dataset, DataLoader

print("train_images shape :", train_images.shape)
print("train_labels shape :", train_labels.shape)
print("val_images shape   :", val_images.shape)
print("val_labels shape   :", val_labels.shape)
print("test_images shape  :", test_images.shape)
print("test_labels shape  :", test_labels.shape)

class ChestMNISTLocalDataset(Dataset):
    def __init__(self, images, labels):
        self.images = images
        self.labels = labels.astype("float32")

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        image = self.images[idx].astype("float32") / 255.0
        label = self.labels[idx]

        image = torch.tensor(image, dtype=torch.float32).unsqueeze(0)
        label = torch.tensor(label, dtype=torch.float32)

        return image, label

train_dataset_torch = ChestMNISTLocalDataset(train_images, train_labels)
val_dataset_torch = ChestMNISTLocalDataset(val_images, val_labels)
test_dataset_torch = ChestMNISTLocalDataset(test_images, test_labels)

batch_size = 128

train_loader = DataLoader(train_dataset_torch, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_dataset_torch, batch_size=batch_size, shuffle=False)
test_loader = DataLoader(test_dataset_torch, batch_size=batch_size, shuffle=False)

print("\nDataLoaders créés avec succès.")
print("Batch size :", batch_size)

sample_images, sample_labels = next(iter(train_loader))
print("Shape batch images :", sample_images.shape)
print("Shape batch labels :", sample_labels.shape)
```

### Préparation des données pour PyTorch

Après avoir chargé les images et les labels depuis le fichier local `chestmnist_128.npz`, les données sont converties dans un format compatible avec PyTorch. Une classe personnalisée de type `Dataset` est créée afin de gérer proprement l’accès aux images et à leurs annotations.

Chaque image est convertie en tenseur et normalisée entre 0 et 1 en divisant les valeurs de pixels par 255. Une dimension supplémentaire est ajoutée pour représenter le canal, ce qui donne un format `(1, 128, 128)` adapté à un réseau de neurones convolutif.

Les jeux d’entraînement, de validation et de test sont ensuite encapsulés dans des `DataLoader`. Cela permet de traiter les données par lots de 128 images, ce qui rend l’entraînement plus efficace et plus stable. Cette étape prépare donc les données pour la phase de modélisation.

```python
import torch.nn as nn

class ChestCNN(nn.Module):
    def __init__(self, num_classes=14):
        super().__init__()

        self.features = nn.Sequential(
            nn.Conv2d(1, 32, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),

            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),

            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2)
        )

        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(128 * 16 * 16, 256),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(256, num_classes)
        )

    def forward(self, x):
        x = self.features(x)
        x = self.classifier(x)
        return x

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model = ChestCNN(num_classes=14).to(device)

criterion = nn.BCEWithLogitsLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

print("Device utilisé :", device)
print(model)
```

### Construction du modèle de classification multilabel

Pour cette première approche, un réseau de neurones convolutif (CNN) est utilisé. Ce choix est cohérent avec la nature des données, car les CNN sont particulièrement adaptés au traitement d’images. Le modèle prend en entrée une radiographie en niveaux de gris de taille `(1, 128, 128)` et extrait progressivement des caractéristiques visuelles à l’aide de couches de convolution, d’activations ReLU et de couches de sous-échantillonnage.

La partie finale du réseau transforme ensuite ces caractéristiques en 14 scores de sortie, correspondant aux 14 pathologies possibles du dataset ChestMNIST. Comme une même image peut contenir plusieurs pathologies simultanément, la tâche est une classification multilabel. C’est pourquoi la fonction de perte choisie est `BCEWithLogitsLoss`, qui est adaptée à ce type de problème.

```python
def train_one_epoch(model, loader, criterion, optimizer, device):
    model.train()
    running_loss = 0.0

    for images, labels in loader:
        images = images.to(device)
        labels = labels.to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    return running_loss / len(loader)

def evaluate(model, loader, criterion, device):
    model.eval()
    running_loss = 0.0

    with torch.no_grad():
        for images, labels in loader:
            images = images.to(device)
            labels = labels.to(device)

            outputs = model(images)
            loss = criterion(outputs, labels)

            running_loss += loss.item()

    return running_loss / len(loader)

num_epochs = 5

train_losses = []
val_losses = []

for epoch in range(num_epochs):
    train_loss = train_one_epoch(model, train_loader, criterion, optimizer, device)
    val_loss = evaluate(model, val_loader, criterion, device)

    train_losses.append(train_loss)
    val_losses.append(val_loss)

    print(f"Epoch [{epoch+1}/{num_epochs}] - Train Loss: {train_loss:.4f} - Val Loss: {val_loss:.4f}")
```

### Entraînement et validation du réseau

Le modèle est ensuite entraîné sur le jeu d’entraînement pendant plusieurs époques. À chaque itération, les images sont transmises au réseau, une prédiction est produite pour les 14 pathologies, puis la perte est calculée à l’aide de la fonction `BCEWithLogitsLoss`. Cette perte est ensuite rétropropagée afin de mettre à jour les poids du modèle avec l’optimiseur Adam.

En parallèle, une phase de validation est réalisée à la fin de chaque époque sur un jeu de données distinct. Cela permet de suivre la capacité du modèle à généraliser sur des données non vues pendant l’apprentissage. L’évolution conjointe de la loss d’entraînement et de validation permet d’évaluer si le modèle apprend correctement ou s’il commence à surapprendre.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.plot(range(1, num_epochs + 1), train_losses, marker='o', label='Train Loss')
plt.plot(range(1, num_epochs + 1), val_losses, marker='o', label='Validation Loss')
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Évolution de la loss d'entraînement et de validation")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import torch
import torch.nn as nn
import copy
import time
import os

class ChestCNNWeighted(nn.Module):
    def __init__(self, num_classes=14):
        super().__init__()

        self.features = nn.Sequential(
            nn.Conv2d(1, 32, kernel_size=3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.MaxPool2d(2),

            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d(2),

            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.MaxPool2d(2)
        )

        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(128 * 16 * 16, 256),
            nn.ReLU(),
            nn.Dropout(0.4),
            nn.Linear(256, num_classes)
        )

    def forward(self, x):
        x = self.features(x)
        x = self.classifier(x)
        return x

device = torch.device("cpu")
weighted_model = ChestCNNWeighted(num_classes=14).to(device)

criterion_weighted = nn.BCEWithLogitsLoss(pos_weight=pos_weight_tensor.to(device))
optimizer_weighted = torch.optim.Adam(weighted_model.parameters(), lr=0.001, weight_decay=1e-4)

scheduler_weighted = torch.optim.lr_scheduler.ReduceLROnPlateau(
    optimizer_weighted,
    mode="min",
    factor=0.5,
    patience=2
)

print(weighted_model)
print("\nLoss pondérée configurée avec succès.")
```

```python
def train_one_epoch(model, loader, criterion, optimizer, device):
    model.train()
    running_loss = 0.0

    for images, labels in loader:
        images = images.to(device)
        labels = labels.to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    return running_loss / len(loader)

def evaluate(model, loader, criterion, device):
    model.eval()
    running_loss = 0.0

    with torch.no_grad():
        for images, labels in loader:
            images = images.to(device)
            labels = labels.to(device)

            outputs = model(images)
            loss = criterion(outputs, labels)

            running_loss += loss.item()

    return running_loss / len(loader)

num_epochs_weighted = 15
patience_early_stopping = 5

weighted_train_losses = []
weighted_val_losses = []

best_weighted_val_loss = float("inf")
best_weighted_weights = copy.deepcopy(weighted_model.state_dict())
epochs_without_improvement = 0

start_time = time.time()

for epoch in range(num_epochs_weighted):
    epoch_start = time.time()

    train_loss = train_one_epoch(weighted_model, train_loader, criterion_weighted, optimizer_weighted, device)
    val_loss = evaluate(weighted_model, val_loader, criterion_weighted, device)

    weighted_train_losses.append(train_loss)
    weighted_val_losses.append(val_loss)

    scheduler_weighted.step(val_loss)
    current_lr = optimizer_weighted.param_groups[0]["lr"]

    print(f"Epoch [{epoch+1}/{num_epochs_weighted}] - "
          f"Train Loss: {train_loss:.4f} - "
          f"Val Loss: {val_loss:.4f} - "
          f"LR: {current_lr:.6f} - "
          f"Temps: {time.time() - epoch_start:.1f}s")

    if val_loss < best_weighted_val_loss:
        best_weighted_val_loss = val_loss
        best_weighted_weights = copy.deepcopy(weighted_model.state_dict())
        epochs_without_improvement = 0
        print("  -> Nouveau meilleur modèle pondéré sauvegardé.")
    else:
        epochs_without_improvement += 1
        print(f"  -> Pas d'amélioration depuis {epochs_without_improvement} époque(s).")

    if epochs_without_improvement >= patience_early_stopping:
        print("\nEarly stopping déclenché.")
        break

total_weighted_time = time.time() - start_time

weighted_model.load_state_dict(best_weighted_weights)

weighted_model_path = os.path.join(dataset_dir, "best_chestcnn_weighted_cpu.pth")
torch.save(weighted_model.state_dict(), weighted_model_path)

print(f"\nMeilleure validation loss (pondérée) : {best_weighted_val_loss:.4f}")
print(f"Temps total d'entraînement : {total_weighted_time/60:.2f} minutes")
print("Modèle pondéré sauvegardé ici :", weighted_model_path)
```

Il n’existe donc pas, parmi les seuils testés ici, de compromis satisfaisant pour ce modèle pondéré. Cela suggère que la pondération utilisée est probablement trop forte, en particulier pour les classes les plus rares, ce qui déforme trop les sorties du réseau.
